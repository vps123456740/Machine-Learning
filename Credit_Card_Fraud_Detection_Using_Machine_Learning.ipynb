Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ccd=pd.read_csv('/content/creditcard.csv')
```

```python
ccd.info()
```

```python
ccd.describe()
```

```python
ccd.isnull().sum()
```

Dropping the missing values


```python


ccd.dropna(inplace=True)
```

```python
ccd.isnull().sum()
```

```python
ccd['Class'].value_counts()
```

This Dataset is highly unbalanced

0: Normal Transaction
1: Fraudulent Transaction

```python
legit=ccd[ccd.Class==0]
fraud=ccd[ccd.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
ccd.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions : 177

```python
legit_sample=legit.sample(n=177)
```

Concatenating two DataFrames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X,Y)
```

Splitting the new data into training data and testing data


```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=5)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy Score on train Data:',training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy Score on test Data:',testing_data_accuracy)
```
